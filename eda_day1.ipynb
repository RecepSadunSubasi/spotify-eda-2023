Streams değişkeni sayısal tipe dönüştürüldükten sonra, en çok dinlenen şarkılar doğru şekilde listelenmiştir.

```python
df.groupby("in_spotify_playlists")["streams"].mean() \
  .sort_index()
```

Spotify playlist sayısı ile ortalama stream sayısı arasında
düzgün ve sürekli bir artış ilişkisi bulunmamaktadır.
Bununla birlikte, yüksek playlist sayısına sahip şarkıların
ortalama stream değerlerinin genel olarak daha yüksek olduğu
gözlemlenmektedir.


```python
df["playlist_group"] = pd.cut(
    df["in_spotify_playlists"],
    bins=[-1, 0, 10, 50, 200, df["in_spotify_playlists"].max()],
    labels=["0", "1–10", "11–50", "51–200", "200+"]
)

df.groupby("playlist_group")["streams"].mean()
```

```python
df[df["playlist_group"] == "0"][["track_name", "streams"]].head()
df[df["playlist_group"] == "1–10"][["track_name", "streams"]].head()
```

Playlist gruplarına göre yapılan analizde,
0 ve 1–10 aralığında playlist sayısına sahip
şarkılar için yeterli gözlem bulunmadığı görülmektedir.

Buna karşın 11–50 ve üzeri playlist gruplarında
ortalama stream değerleri belirgin şekilde artmaktadır.
Bu durum, yüksek playlist görünürlüğünün
şarkı popülerliği ile ilişkili olduğunu göstermektedir.

"0 ve 1–10 playlist gruplarında yer alan şarkılara
veri setinde rastlanmamıştır.
Bu nedenle bu gruplar için ortalama stream analizi
yapılamamıştır."
